```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
con = requests.get(url).content
```

```python
content = BeautifulSoup(con, 'html.parser')
```

```python
content
```

```python
table_content = content.find('tbody').find_all("tr")
```

```python
table_content
```

```python
table_content[8]
```

```python
for i in range (8, len(table_content)):
    print(table_content[i])
   
```

```python
all_data = []

for i in range (8, len(table_content)):
    dat = table_content[i]
    country_row = []
    td = dat.find_all('td')
    for item in td:
        country_row.append(item.text)
    all_data.append(country_row)
```

```python
all_data
```

```python
covid_data = list(map(lambda x:x[1:10]+[x[12], x[14]], all_data ))
```

```python
columns_names = ["Name",
               "Total Cases",
               "New Cases",
               "Total Deaths",
               "New Deaths",
               "Total Recovered",
               "New recovered",
               "Active Cases",
               "Serious Cases",
               "Total Tezts",
               "Population"]
```

```python
covid_data[0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(covid_data, columns = columns_names)
```

```python
df.to_csv('covid-data.csv', index = False)
```
